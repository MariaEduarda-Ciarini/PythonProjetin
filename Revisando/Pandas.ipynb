```python
dados['Gross Value'] = dados ['Valor'] + dados['Condominio'] + dados ['IPTU']
dados.head(10)

dados['Value m²'] = dados['Valor'] / dados ['Area']
dados.head(10)

dados['Value m²'] = dados ['Value m²'].round(2)
dados

dados['Gross Value m²'] = (dados['Gross Value'] / dados ['Area']).round(2)
dados.head(10)
```

```python
house = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
dados['Aggregate Type'] = dados ['Tipo'].apply(lambda x: 'Casa' if x in house else 'Apartamento')
dados
```
