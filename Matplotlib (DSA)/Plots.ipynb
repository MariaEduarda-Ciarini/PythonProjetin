```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
%matplotlib inline
```

```python
# Plot Básico

plt.plot([3, 6, 9, 12, 15, 18], [5, 15, 20, 25, 32, 13])
plt.xlabel('eixo 1')
plt.ylabel('eixo 2')
plt.title("Gráfico")
plt.show()
```

```python
print(plt.style.available) 
```

```python
plt.style.use('dark_background')
x1 = [0.5, 5, 9.99, 12.99]
y2 = [0, 10, 15, 20]

plt.plot(x1, y2, label="Matplotlib", c='purple')
plt.legend()
plt.show()
```

```python
# Plot Bar
mpl.rcParams.update(mpl.rcParamsDefault)

x3 = [4, 6, 8, 10, 2]
y3 = [6, 7, 8, 2, 4]

plt.bar(x3, y3, label = "Gráfico de Barras", color = 'red')
plt.legend()
plt.show()
```

```python
x4 = [1, 3, 5, 7, 9]
y4 = [7, 8, 2, 4, 2]

plt.bar(x3, y3, label='Lista1', color='blue')
plt.bar(x4, y4, label='Lista2', color='red')
plt.legend()
plt.show()
```
