```python
from pylab import *
%matplotlib inline
```

```python
# Plot Line

x = linspace(0, 5, 10)
y = x ** 2

fig = plt.figure()

axes = fig.add_axes([0, 0, 0.8, 0.8])
axes.plot(x, y, 'g')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Line Plot')
```

```python
# Dados
x = linspace(0, 5, 10)
y = x ** 2

# Cria figura
fig = plt.figure()

# Cria os eixos
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # Eixos da figura 1
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])  # Eixos da figura 2

# Figura Principal
axes1.plot(x, y, 'purple')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Main Figure')

# Figura Secund√°ria
axes2.plot(x, y, 'gray')
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('Second Figure')
```

```python
x = linspace(0, 5, 10)
y = x ** 2

# Divide a √°rea de plotagem em dois subplots

fig, axes = plt.subplots(nrows=1, ncols=2)

# loop pelos eixos para criar cada plot

for ax in axes:
    ax.plot(x, y,  'red')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Plotb')

fig.tight_layout()
```

```python
x = linspace(0, 5, 10)
y = x ** 2

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(x, x**2, x, exp(x))
axes[0].set_title("Escala Padr√£o")

axes[1].plot(x, x**2, x, exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Escala Logaritmica (y)")
```

```python
# Grid

x = linspace(0, 5, 10)
y = x ** 2

fig, axes = plt.subplots(1, 2, figsize=(10, 3))

# Grid Padr√£o
axes[0].plot(x, x**2, x, x**3, lw=2)
axes[0].grid(True)

# Grid Custumizado
axes[1].plot(x, x**2, x, x**3, lw=1)
axes[1].grid(color='blue', alpha=0.7, linestyle='dashed', linewidth=0.8)
```

```python
# Diferentes estilos de Plots

# Dados
xx = np.linspace(-0.75, 1, 100)
n = np.array([0, 1, 2, 3, 4, 5])

# Subplots
fig, axes = plt.subplots(1, 4, figsize=(12, 3))

# Plot 1
axes[0].scatter(xx, xx + 0.25 * randn(len(xx)), color="red")
axes[0].set_title("scatter")

# Plot 2
axes[1].step(n, n**2, lw=2, color="orange")
axes[1].set_title("step")

# Plot 3
axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5, color="purple")
axes[2].set_title("bar")

# Plot 4
axes[3].fill_between(x, x ** 2, x**3, alpha=0.5, color="green")
axes[3].set_title("fill_between")
```
