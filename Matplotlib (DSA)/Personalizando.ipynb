```python
ano =[1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]

dado =[155, 186,  38,  59, 159, 197,  74,  28, 105,  16,  59, 172,  66, 110,  27,  51, 105,  10,
7, 166,  68,  89, 169, 193,  58,  46, 108,  86, 124, 144, 200,  66,  80,129]
```

```python
plt.plot(ano, dado, color='blue', marker='o', markersize=4)
plt.title("Testando")
plt.xlabel("ano")
plt.ylabel("dados random")
plt.tight_layout()
plt.show()
```

```python
plt.plot(ano, dado, color='blue', marker='o',
         linestyle='dotted', linewidth=1.50)
plt.title("Testando")
plt.xlabel("ano")
plt.ylabel("dados random")
plt.tight_layout()
plt.show()
```

```python
plt.fill_between(ano, dado, 0, color='orange', alpha=0.40)
plt.title("Testando")
plt.xlabel("ano")
plt.ylabel("dados random")
plt.tight_layout()
plt.show()
```

```python
plt.plot(ano, dado)
plt.scatter(ano, dado, color='red', marker='o', alpha=1.0)
plt.title("Testando")
plt.xlabel("ano")
plt.ylabel("dados random")
plt.tight_layout()
plt.show()
```
