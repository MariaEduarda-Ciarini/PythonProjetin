```python
import matplotlib.pyplot as plt
import matplotlib as mpl


# Plot Scatter

x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [5, 2, 4, 5, 6, 8, 4, 8]

plt.scatter(x, y, label = 'Scatter', color = 'yellow', marker = "*")
plt.legend()
plt.show()
```

```python
# Plot Pizza

labels = 'Cachorros', 'Gatos', 'Coelhos', 'Louros'
tamanhos = [35, 30, 15, 20]

fig, ax = plt.subplots()
ax.pie(tamanhos, labels=labels)

plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(tamanhos, labels=labels, autopct='%1.1f%%')
```

```python
fig, ax = plt.subplots()
ax.pie(tamanhos, labels=labels, colors=['brown', 'gray', 'pink', 'green'])
```

```python
fig, ax = plt.subplots()
ax.pie(tamanhos, labels=labels, autopct='%1.1f%%',
       pctdistance=1.20, labeldistance=.7)
```

```python
explode = (0, 0.1, 0, 0)

fig, ax = plt.subplots()
ax.pie(tamanhos, explode=explode, labels=labels,
       autopct='%1.1f%%', shadow=True, startangle=90)
plt.show()
```
