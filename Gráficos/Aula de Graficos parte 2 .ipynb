# stackplot 100%

```python
mpl.rcParams.update(mpl.rcParamsDefault)
plt.style.use('fivethirtyeight')
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 11
fig_size[1] = 3

labels = ["PIB - Agricultura", "PIB  - Industria", "PIB - Serviços"]
colors = "green", "black", "blue"
plt.stackplot(ano, pib_agro_perc, pib_ind_perc,
              pib_serv_perc, labels=labels, colors=colors)

plt.legend(loc=(0.07, 0.50))
plt.title("Evolução do PIB do Brasil - Composição Setorial Percentual")
plt.tight_layout()


plt.show()
```
