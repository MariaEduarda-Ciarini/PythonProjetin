# Com dados reais 
# Origem dos dados em arquivo csv.

```python
import pandas as pd
import csv
```

```python
data = pd.read_csv('data_TrendsYoutube.csv')
view_count=data ['view_count']
likes=data ['likes']
ratio= data ['ratio']
```

```python
plt.style.use('ggplot')

fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 8
fig_size[1] = 4

plt.scatter(view_count, likes, c=ratio, cmap='summer',
            edgecolor='black', linewidth=1, alpha=0.85)

cbar = plt.colorbar()
cbar.set_label('Razão Like/Dislike')

plt.title("Comportamento de Vídeos no Youtube")
plt.xlabel("Visualizações")
plt.ylabel("Likes")

plt.tight_layout()
plt.show()
```

```python
plt.style.use('ggplot')

fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 8
fig_size[1] = 4

plt.scatter(view_count, likes, c=ratio, cmap='summer',
            edgecolor='black', linewidth=1, alpha=0.85)

cbar = plt.colorbar()
cbar.set_label('Razão Like/Dislike')

plt.xscale('log')
plt.yscale('log')

plt.title("Comportamento de Vídeos no Youtube")
plt.xlabel("Visualizações")
plt.ylabel("Likes")

plt.tight_layout()
plt.show()
```
