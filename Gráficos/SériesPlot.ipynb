# Gráficos de Séries Temporais e suas Flexibilizações

```python
import matplotlib as mpl
from matplotlib import pyplot as plt
from datetime import datetime
```

```python
dates = [
datetime(2021, 7, 20),
datetime(2021, 7, 21),
datetime(2021, 7, 22),
datetime(2021, 7, 23 ),
datetime(2021, 7, 24),
datetime(2021, 7, 25),
datetime(2021, 7, 26)
]

dates

y =[2, 3, 4, 1.5, 5, 6, 7]
```

```python
plt.style.use("ggplot")
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 8
fig_size[1] = 4

plt.plot_date(dates, y, linestyle='-')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
plt.style.use("ggplot")

fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 8
fig_size[1] = 4

plt.plot_date(dates, y, linestyle='-')

plt.gcf().autofmt_xdate()

plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
from matplotlib import dates as mpl_dates

plt.style.use("ggplot")

fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 8
fig_size[1] = 4

plt.plot_date(dates, y, linestyle='solid')

plt.gcf().autofmt_xdate()

date_format = mpl_dates.DateFormatter('%b, %d %Y')

plt.gca().xaxis.set_major_formatter(date_format)

plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import csv
```

```python
data = pd.read_csv('data_bitcoinPrice.csv')
data
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date')
price_date = data['Date']
price_close = data['Close']
```

```python
from matplotlib import dates as mpl_dates

plt.style.use("ggplot")

fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 8
fig_size[1] = 4

plt.plot_date(price_date, price_close, linestyle='solid')
plt.gcf().autofmt_xdate()


plt.title('Preços de Bitcoin')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')

plt.tight_layout()
plt.grid(True)
plt.show()
```
